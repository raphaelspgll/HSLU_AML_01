# Dataset Overview: Capital Bikeshare (Hourly)

This dataset contains the hourly count of rental bikes from the **Capital Bikeshare system** in Washington D.C. The data is heavily correlated with environmental, seasonal, and temporal settings.

### Metadata
- **Source:** Real-world data extracted from the Capital Bikeshare system.
- **Size:** $N = 17,379$ observations (using the `hour.csv` file).
- **Predictors:** 16 variables.

---

### Variable Type Breakdown

| Variable Type | Variables | Description |
| :--- | :--- | :--- |
| **Continuous** | `temp`, `hum`, `windspeed` | Normalized temperature, humidity, and wind speed. |
| **Categorical** | `season`, `weathersit` | **Season:** (1:spring, 2:summer, 3:fall, 4:winter). <br> **Weather:** (4 levels from Clear to Heavy Rain/Snow). |
| **Count** | `casual`, `registered`, `cnt` | Number of non-registered users, registered users, and total rentals. |
| **Binary** | `workingday`, `holiday` | (1) if the day is a working day/holiday, (0) otherwise. |



---

### Requirements
- [x] **Moderate Size:** $1.7 \times 10^4$ observations (within $[10^3, 10^5]$).
- [x] **Real Data:** Yes.
- [x] **Mixed Types:** Includes Continuous, Categorical, Count, and Binary.
- [x] **Multi-level Categorical:** `season` and `weathersit` both have 4 levels.

## Dataset Characteristics

Both hour.csv and day.csv have the following fields, except hr which is not available in day.csv
	
	- instant: record index
	- dteday : date
	- season : season (1:springer, 2:summer, 3:fall, 4:winter)
	- yr : year (0: 2011, 1:2012)
	- mnth : month ( 1 to 12)
	- hr : hour (0 to 23)
	- holiday : weather day is holiday or not (extracted from http://dchr.dc.gov/page/holiday-schedule)
	- weekday : day of the week
	- workingday : if day is neither weekend nor holiday is 1, otherwise is 0.
	+ weathersit : 
		- 1: Clear, Few clouds, Partly cloudy, Partly cloudy
		- 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
		- 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
		- 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog
	- temp : Normalized temperature in Celsius. The values are divided to 41 (max)
	- atemp: Normalized feeling temperature in Celsius. The values are divided to 50 (max)
	- hum: Normalized humidity. The values are divided to 100 (max)
	- windspeed: Normalized wind speed. The values are divided to 67 (max)
	- casual: count of casual users
	- registered: count of registered users
	- cnt: count of total rental bikes including both casual and registered


```python
def data_cleaning_helper(data: pd.DataFrame):
    print(data.isnull().sum())
    return data.drop(columns=['instant', 'dteday', 'atemp'])
```

# Dataset Overview: Bank Marketing (Full)

This dataset contains information related to direct marketing campaigns (phone calls) of a Portuguese banking institution. The classification goal is to predict if the client will subscribe to a term deposit.

### Metadata
- **Source:** Real-world data from a Portuguese retail bank (UCI Machine Learning Repository).
- **Size:** $N = 45,211$ observations (using the `bank-full.csv` file).
- **Predictors:** 16 variables (plus the target variable).

---

### Variable Type Breakdown

| Variable Type | Variables | Description |
| :--- | :--- | :--- |
| **Continuous** | `balance`, `duration` | **Balance:** Average yearly balance in euros. <br> **Duration:** Last contact duration in seconds. |
| **Categorical** | `job`, `education`, `month` | **Job:** (12 levels: admin, blue-collar, etc.). <br> **Education:** (primary, secondary, tertiary, unknown). |
| **Count** | `campaign`, `pdays`, `previous` | **Campaign:** Number of contacts during this campaign. <br> **Previous:** Number of contacts before this campaign. |
| **Binary** | `housing`, `loan`, `default`, `y` | **Housing/Loan:** (yes/no) status. <br> **y:** (Target) Has the client subscribed? (yes/no). |

---

### Requirements
- [x] **Moderate Size:** $4.5 \times 10^4$ observations (within $[10^3, 10^5]$).
- [x] **Real Data:** Yes.
- [x] **Mixed Types:** Includes Continuous, Categorical, Count, and Binary.
- [x] **Multi-level Categorical:** `job` (12 levels) and `education` (4 levels) meet the requirement.
